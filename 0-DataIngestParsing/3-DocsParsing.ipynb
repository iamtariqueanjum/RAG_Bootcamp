### Word Document Processing

### Docx2txtLoader

```python
from langchain_community.document_loaders import Docx2txtLoader

try:
    doc_loader = Docx2txtLoader(file_path="data/word_files/proposal.docx")
    documents = doc_loader.load()
    print(f"Loaded {len(documents)} documents")
    print(f"First document page_content :{documents[0].page_content[:200]}")
    print(f"First document metadata :{documents[0].metadata}")
except Exception as e:
    print(f"Error : {e}")
```

### UnstructuredWordDocumentLoader

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

try:
    doc_loader = UnstructuredWordDocumentLoader(file_path="data/word_files/proposal.docx", mode="elements")
    # mode i.e, elements splits it into different documents by element
    documents = doc_loader.load()
    print(f"Loaded {len(documents)} documents")
    print(f"First document page_content :{documents[0].page_content[:200]}")
    print(f"First document metadata :{documents[0].metadata}")
except Exception as e:
    print(f"Error : {e}")
```
